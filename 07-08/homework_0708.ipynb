```python
import pandas as pd
```

```python
# 为了便于操作，已将csv文件通过文本编辑器修改了编码格式，改为utf-8
source = pd.read_csv(r"C:\Users\zhang\Desktop\AI_and_NLP_homework\07-08\sqlResult_1558435.csv")
```

```python
source.head(3)
```

```python
base_df = source[['source', 'title', 'content']]
```

```python
base_df.head(5)
```

```python
base_df['labels'] = 0
base_df.loc[base_df['source'] == '新华社', 'labels'] = 1
base_df = base_df.dropna()
```

```python
base_df.head(5)
```

```python
from sklearn.utils import shuffle
```

```python
df_shuffled = shuffle(base_df).reset_index()
```

```python
news_contents = df_shuffled['content'].tolist()
news_targets = df_shuffled['labels'].tolist()
news_source = df_shuffled['source'].tolist()
```

```python
l = len(news_contents)
```

```python
import jieba
```

```python
stop_words = []
with open(r'C:\Users\zhang\Desktop\AI_and_NLP_homework\07-08\chinese_stopwords.txt', mode='r', encoding='utf8') as f:
    for word in f.readlines():
        stop_words.append(word.strip())
```

```python
for i in range(l):
    content = news_contents[i]
    s = []
    for w in jieba.cut(content):
        if (w in stop_words) or (w == '\\') or (w == 'n'):
            continue
        s.append(w.strip())
    if i % 5000 == 0:
        print(i)
    news_contents[i] = ' '.join(s)
```

# 文本整理完毕，开始使用词袋将文本表征成矩阵

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec = CountVectorizer()
```

```python
X_train_counts = count_vec.fit_transform(news_contents)
```

```python
X_train_counts.shape
```

```python
count_vec.vocabulary_
```

# 使用tf-idf将词出现的次数变成频率

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf_transformer = TfidfTransformer()
```

```python
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
```

```python
X_train_tfidf.shape
```

# 使用机器学习算法分类

```python
from sklearn.model_selection import train_test_split
```

```python
contents_train, contents_test, targets_train, target_test = train_test_split(X_train_tfidf, news_targets)
```

```python
# 朴素贝叶斯
from sklearn.naive_bayes import MultinomialNB
```

```python
model_nb = MultinomialNB()
```

```python
clf_nb = model_nb.fit(contents_train, targets_train)
```

```python
target_predict = clf_nb.predict(contents_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(target_test, target_predict))
```

```python
print(metrics.confusion_matrix(target_test, target_predict))
```

```python
metrics.roc_auc_score(target_test, target_predict)
```

### 根据sklearn中混淆矩阵的排列方式，可以得到在朴素贝叶斯算法下：
### 为新华社文章:
### precision = 19665/(2925+19665) = 0.91
### recall = 19665/(19665+2) = 1
### f1 = 2*(precision*recall)/(precision+recall) = 0.95
