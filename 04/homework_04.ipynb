```python
import json
from jieba import cut
from hanziconv import HanziConv
```

```python
def text_cut(text):
    text = HanziConv.toSimplified(''.join(text.strip().split()))
    text = ' '.join(cut(text))
    return text
```

```python
count = 0
source_path = 'C:\\Users\\zhang\\Desktop\\'
with open(source_path+'corpus.txt', mode='w', encoding='utf8') as f_write:
    for folder in os.listdir(source_path+'results\\'):
        folder_path = source_path + 'results\\' + folder + '\\'
        for file in os.listdir(folder_path):
            file_path = folder_path + file
            with open(file_path, mode='r', encoding='utf8') as f_read:
                for line in f_read.readlines():
                    data = json.loads(line)
                    text = data['text']
                    text = text_cut(text)
                    f_write.write(text + '\n')
                    count += 1
                    if count % 1000 == 0:
                        print(count)
```

```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
```

```python
# 这里临时使用了一个某教材的训练英文语料的参数，但是从结果来看对中文语料的结果并不理想

import multiprocessing
#ONCE we have vectors
#step 3 - build model
#3 main tasks that vectors help with
#DISTANCE, SIMILARITY, RANKING

# Dimensionality of the resulting word vectors.
#more dimensions, more computationally expensive to train
#but also more accurate
#more dimensions = more generalized
num_features = 300
# Minimum word count threshold.
min_word_count = 3

# Number of threads to run in parallel.
#more workers, faster we train
num_workers = multiprocessing.cpu_count()

# Context window length.
context_size = 7

# Downsample setting for frequent words.
#0 - 1e-5 is good for this
downsampling = 1e-3

# Seed for the RNG, to make the results reproducible.
#random number generator
#deterministic, good for debugging
seed = 1
```

```python
wiki2vec = word2vec.Word2Vec(
    sg=1,
    seed=seed,
    workers=num_workers,
    size=num_features,
    min_count=min_word_count,
    window=context_size,
    sample=downsampling
)
```

```python
LineSentence(source_path+'corpus.txt')
```

```python
wiki2vec.build_vocab(corpus_file=source_path+'corpus.txt')
```
